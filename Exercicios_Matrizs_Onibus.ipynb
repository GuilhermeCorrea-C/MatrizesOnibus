```python
def preencheAssentos(matriz):
  continuar = 's'
  while continuar == 's':
    idade = int(input('IDADE DO PASSAGEIRO: '))
    fileira = int(input('FILEIRA DESEJADA: ')) -1
    assento = int(input('ASSENTO DESEJADO: ')) -1 
    if matriz[fileira][assento] != 0:
      print('ASSENTO OCUPADO, FAVOR SELECIONAR OUTRO ASSENTO')
    else:
      matriz[fileira][assento] = idade
    continuar = input('CONTINUAR? S OU N: ')

  return matriz

def assentosOucpLiv(matr):
  livres = 0
  ocupados = 0
  for i in range(len(matr)):
    for j in range(len(matr[i])):
      if matr[i][j] == 0:
        livres += 1
      else: 
        ocupados += 1
  print(f'{livres} assentos livres e {ocupados} assentos ocupados')

def mediaIdades(mat):
  l_idades = []
  cont = 0
  soma = 0
  for i in range(len(mat)):
    for j in range(len(mat[i])):
      if mat[i][j] > 0:
        l_idades.append(mat[i][j])
        cont += 1
  for x in range(len(l_idades)):
    soma += l_idades[x]
  return soma / cont 

def totalArrecadado(mat):
  total = 0
  for i in range(len(mat)):
    for j in range(len(mat[i])):
      if 0 < mat[i][j] <= 12 and i == 1:
        total += 100
      elif 0 < mat[i][j] <= 12:
        total += 150
      elif 13 <= mat[i][j] <= 18 and i== 1:
        total += 200
      elif 13 <= mat[i][j] <= 18:
        total += 260
      elif mat[i][j] >= 19 and i==1:
        total += 300
      elif mat[i][j] >= 19:
        total += 370
          
  return total 

def maiorIdade(mat):
  maior = 0
  for i in mat:
    for j in i:
      if j > maior:
        maior = j

  return maior

def main():
  matrix = [[0]*16 for i in range(3)]
  matriz = preencheAssentos(matrix) 
  print(matriz)
  assentosOucpLiv(matriz)
  media = mediaIdades(matriz)
  print(f'a média é {media}')
  total = totalArrecadado(matriz)
  print(f'o total arrecadado foi de {total} reais')
  maiorI = maiorIdade(matriz)
  print(f'a maior idade cadastrada é de {maiorI} anos')

main()
```
